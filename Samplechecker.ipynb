```python
import os
import matplotlib.pyplot as plt
import numpy as np
cla_obj = ('Smoke Detector',                  'Alarm Button',          'Fire Extinguisher',              'Ceiling Light',
           'FM Power Point (Plug)',       'Command Point (Switch)',          'Security Cameras') 
num_Instance = [0, 0, 0, 0, 0, 0, 0]

dir_root = r"D:\Kai\Kai_work\240703_BG15\Images\342\1st\Annotation\fisheye"

for root, dirs, files in os.walk(dir_root):
    for file in files:
        if file.endswith(".txt"):
            with open(os.path.join(root, file), 'r') as f:
                lines = f.readlines()
                for line in lines:
                    loc = int(line.split(' ')[0])
                    num_Instance[loc] += 1

print(num_Instance)
```

```python
#plot a bar chart
fig, ax = plt.subplots()
y_pos = np.arange(len(cla_obj))
ax.barh(y_pos, num_Instance, align='center')
ax.set_yticks(y_pos)
ax.set_yticklabels(cla_obj)
ax.invert_yaxis()  # labels read top-to-bottom
ax.set_xlabel('Number of Instances')
ax.set_title('Number of Instances for each class')
plt.show()
```
